Eliminando columnas vacias

```python
df = df.dropna()
```

```python
df
```

Transformacion de las columnas de

*   Worldwide
*   Domestic
*   Foreign


Esta parte tranforma los valores a string no ejecutar si se busca hacer operaciones matematicas



```python
df['$Worldwide'] = df['$Worldwide'].astype('Int64')
df['$Worldwide'] = df['$Worldwide'].apply(lambda x: f"{int(x):,}")
```

```python
df['$Domestic'] = df['$Domestic'].astype('Int64')
df['$Domestic'] = df['$Domestic'].apply(lambda x: f"{int(x):,}")
```

```python
df['$Foreign'] = df['$Foreign'].astype('Int64')
df['$Foreign'] = df['$Foreign'].apply(lambda x: f"{int(x):,}")
```

```python
df
```

se agrega simbolos y se eliminan digitos no relevantes y se convierte la categoria de generos en una lista


```python
df['Domestic %'] = df['Domestic %'].astype(float).apply(lambda x: f"{x:.1f}%")
df['Foreign %'] = df['Foreign %'].astype(float).apply(lambda x: f"{x:.1f}%")
df['Vote_Count'] = df['Vote_Count'].astype(int)
df['Genres'] = df['Genres'].astype(str).apply(lambda x: x.split(','))
```
